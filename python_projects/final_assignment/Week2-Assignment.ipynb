# **1) IPL and RCB**

to qualify for playoffs. It is only possible if they win all their 
remaining 5 matches.

In second case X=1 and Y=5, so RCB needs 1 points from their remaining 5 
matches to qualify for playoffs. It can be done if they tie any one of their 5 matches and lose the remaining 4. So they need to win 0 matches

```
# Write a function
def min_number_of_wins_to_playoffs():
    number_of_testcases = int(input())
    
    while number_of_testcases>0:
        
        winning_score,tied_score,losing_score = 2,1,0
        input1,input2 = input().split()
        threshold_score = int(input1)
        matches = int(input2)
        
        min_matches_to_playoffs = 0
        if threshold_score <= 2 * matches:              #matching with conatraint that 0<=X<=2*Y
            while threshold_score >= 0:
                if threshold_score >= winning_score:
                    min_matches_to_playoffs += 1

                threshold_score -= winning_score
        else:                                           # is the X exceeds the 2*Y
            print('invalid input : the minimum score to the playoffs exceeds the maximum limit of the score available')
        
        print(min_matches_to_playoffs)
        
        number_of_testcases -= 1


# Call the function
min_number_of_wins_to_playoffs()
```

# **3) Recaman’s sequence**

Given an integer n.  Write a function to print first n elements of Recaman’s sequence.

a(0) = 0,

if n > 0 and the number a(n-1) - n > 0 and is not 

   already included in the sequence,

     a(n) = a(n - 1) - n 

else 

     a(n) = a(n-1) + n.

<hr>     

Input:

Integer n>0

Output:

Array of Recaman's Sequence in ascending order.

Examples:

Input : n = 6

Output : {[0, 1, 3, 6, 2, 7]}

Input  : n = 17

Output : {[0, 1, 3, 6, 2, 7, 13, 20, 12, 21, 
         11, 22, 10, 23, 9, 24, 8]}
<br><br>
Hint: Use Recursion         

```
# import necessary modules
# Write a recursive function taking n as input.
def recaman_series(n):
    a = [0]*n
    series = []
    a[0] = 0 #by observation a(0) is always 0
    series.append(a[0])

    for num in range(1,n):
        a_of_num = a[num-1] - num

        for check in range(0,num):
            if ((a_of_num == a[check]) or a_of_num < 0):
                a_of_num = a[num-1] + num
                break

        a[num] = a_of_num
        series.append(a[num])

    return series

# Call the function.
n = int(input())
print(recaman_series(n))

# You can add anything else to this code if you feel it is required.
```

Base is the shortest side of the triangle

```
# You Can Do It !!!
def number_of_squares(n):       #function for calculating the required number of squares.
    return (n*(n+1))//2

no_of_cases = int(input())      #number of test_cases

while no_of_cases > 0:
    B = int(input())            #taking the length of base as B
    n = (B//2)-1
    print(number_of_squares(n)) # calling of function number_of_squares 
    no_of_cases -= 1
```
